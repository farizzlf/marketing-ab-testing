```python
import scipy.stats as stats

# Parameters
alpha = 0.05
power = 0.8
std_dev = 0.1
delta = 0.01

# Calculate critical values
z_alpha = stats.norm.ppf(1 - alpha / 2)
z_beta = stats.norm.ppf(power)

# Calculate sample size per group
sample_size_per_group = ((2 * (std_dev**2) * (z_alpha + z_beta)**2) / delta**2)
sample_size_per_group = round(sample_size_per_group)

sample_size_per_group
```

```python
import pandas as pd

# Mengambil data dari file CSV setelah sampling
file_path = '/content/marketing_AB_sampiling.csv'
df = pd.read_csv(file_path)

# Menghitung jumlah total pengguna di masing-masing grup
total_ad_users = df[df['test group'] == 'ad'].shape[0]
total_psa_users = df[df['test group'] == 'psa'].shape[0]

# Menghitung jumlah konversi di masing-masing grup
converted_ad = df[(df['test group'] == 'ad') & (df['converted'] == True)].shape[0]
converted_psa = df[(df['test group'] == 'psa') & (df['converted'] == True)].shape[0]

# Menghitung persentase konversi untuk masing-masing grup
conversion_rate_ad = (converted_ad / total_ad_users) * 100
conversion_rate_psa = (converted_psa / total_psa_users) * 100

# Menampilkan hasil
print(f"Conversion rate for ad group: {conversion_rate_ad:.2f}%")
print(f"Conversion rate for PSA group: {conversion_rate_psa:.2f}%")
```

```python
import pandas as pd
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Mengambil data dari file CSV setelah sampling
file_path = '/content/marketing_AB_sampiling.csv'
df = pd.read_csv(file_path)

# Memisahkan data untuk masing-masing grup
group_ad = df[df['test group'] == 'ad']
group_psa = df[df['test group'] == 'psa']

# Menghitung jumlah pengguna dan konversi di masing-masing grup
converted_ad = group_ad['converted'].sum()
total_ad = group_ad.shape[0]

converted_psa = group_psa['converted'].sum()
total_psa = group_psa.shape[0]

# Menghitung proporsi konversi untuk masing-masing grup
prop_ad = converted_ad / total_ad
prop_psa = converted_psa / total_psa

# Menampilkan informasi tentang jumlah dan konversi masing-masing grup
print(f"Total users in ad group: {total_ad}")
print(f"Converted users in ad group: {converted_ad}")
print(f"Total users in PSA group: {total_psa}")
print(f"Converted users in PSA group: {converted_psa}")

# Melakukan uji proporsi Z untuk A/B testing
count = np.array([converted_ad, converted_psa])
nobs = np.array([total_ad, total_psa])

z_score, p_value = proportions_ztest(count, nobs)

# Menampilkan hasil uji statistik
print(f"Z Score: {z_score:.4f}")
print(f"P-Value: {p_value:.4f}")

# Menentukan hasil berdasarkan p-value
alpha = 0.05
if p_value < alpha:
    print("P-Value is less than alpha. Reject null hypothesis.")
    print("There is a significant difference in conversion rates between ad and PSA groups.")
else:
    print("P-Value is greater than or equal to alpha. Fail to reject null hypothesis.")
    print("There is no significant difference in conversion rates between ad and PSA groups.")
```
